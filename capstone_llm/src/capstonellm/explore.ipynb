```python
s3 = boto3.client('s3')
bucket='dataminded-academy-capstone-llm-data-us'

result = s3.list_objects(Bucket = bucket, Prefix='input')
tags = set()
for o in result.get('Contents'):
    temp_key = o['Key']
    print(temp_key)
    temp_tag = temp_key.split('/')[1]
    temp_file_name = temp_key.replace('/','_')
    tags.add(temp_tag)
    s3.download_file(bucket, temp_key, f'../../data/{temp_file_name}')
print(tags)
```

```python
tag = 'dbt'
dfs_input = {}
for tag in tags:
    dfs_input[f'input_{tag}_questions'] = spark.read.json(f'../../data/input_{tag}_questions.json').select(fn.explode(fn.col('items'))).select('col.*')
    dfs_input[f'input_{tag}_answers'] = spark.read.json(f'../../data/input_{tag}_answers.json').select(fn.explode(fn.col('items'))).select('col.*')

for name, df in dfs_input.items():
    print(name)
    df.show(1,False) 
```

```python
for tag in tags:
    q = dfs_input[f'input_{tag}_questions'].alias('question')
    a = dfs_input[f'input_{tag}_answers'].alias('answer')
    r = (
        q
        .join(a, [fn.col('question.accepted_answer_id') == fn.col('answer.answer_id')], 'left')
        .select(fn.col('question.title').alias('title'), fn.col('question.body').alias('question_body'), fn.col('answer.body').alias('response_body'))
        )
    r.show(1)
    r.repartition(r.count()).write.json(f'../../data/cleaned_{tag}')
```
